```python
import pandas as pd  
import altair as alt
```

```python
holiday = pd.read_csv('buddymove_holidayiq.csv')

holiday
```

```python
dashboard1 = alt.Chart(holiday).mark_bar().encode(
    x = 'Sports',
    y = 'Religious'
)
dashboard1
```

```python
dashboard2 = alt.Chart(holiday).mark_bar().encode(
    x = 'Sports', 
    y = 'Nature'
)
dashboard2
```

### Table Key
User id = Unique user id

Sports: Number of reviews on stadiums, sports complex, etc.

Religious: Number of reviews on religious institutions

Nature: Number of reviews on beach, lake, river, etc.

Theatre: Number of reviews on theatres, exhibitions, etc.

Shopping : Number of reviews on malls, shopping places, etc.

Picnic : Number of reviews on parks, picnic spots, etc.



### Decisions made and why

A bar chart was used as this gave the best display of a potential correlation between the variables used from the dataset.

### Data insights
Generally there is a positive correlation between the amount of reviews given on each category of facilities on holiday destinations. This means the more reviews a user gave on sports facilities the more that same user gave on other categories such as nature or religious with the exception of a few anomalies. It is safe to assume the activity level of each user based on the amount of reviews given per category. The more active a person is on the holiday to more reviews where given per category.
